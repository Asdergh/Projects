Функция сортировки данных

```python
def list_sorter(list_to_sort):
    for i in range(len(list_to_sort)):
        for j in range(i, 0, -1):
            if list_to_sort[j] > list_to_sort[i]:
                list_to_sort[j], list_to_sort[i] = list_to_sort[i], list_to_sort[j]
    return list_to_sort
```

Разбор данных

```python
with open("data.json", "r") as file:
    data = js.load(file)
    need_data_height = []
    need_data_weight = []
    names = []
    fig, axes_2d = plt.subplots()
    for elem in range(1, len(data)):
        need_data_height.append((data[str(elem)]["Height"]))
        names.append(data[str(elem)]["Name"])
        need_data_weight.append(data[str(elem)]["Width"])

    x_grid = [x for x in range(len(names))]

    axes_2d.scatter(x_grid, need_data_height, color="green", s=20)
    axes_2d.plot(x_grid, need_data_height, color="red")

    axes_2d.scatter(x_grid, need_data_weight, color="red")
    axes_2d.plot(x_grid, need_data_weight, color="green")

    need_data_height = list_sorter(need_data_height)
    need_data_weight = list_sorter(need_data_weight)

    axes_2d.scatter(x_grid, need_data_height, color="black")
    axes_2d.plot(x_grid, need_data_height, color="red", label="данные по высоте учатсников")

    axes_2d.scatter(x_grid, need_data_weight, color="m")
    axes_2d.plot(x_grid, need_data_weight, color="green", label="данные по весу участников")

    axes_2d.legend(loc="upper left")
    data_dispersion_height = stat.variance(need_data_height)
    data_dispersion_weight = stat.variance(need_data_weight)

    


    print(f"Height dispersion: {data_dispersion_height}, Weight dispersion: {data_dispersion_weight}")
    plt.xticks(range(len(names)), names, rotation=45)
    plt.show()

    
        
        
```
